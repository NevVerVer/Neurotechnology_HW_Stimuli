```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# PATH = /trial.csv
df = pd.read_csv('/trial.csv')
df[:5]
```

```python
q = df['key timestamp'] + np.random.uniform(low=-0.1, high=0.75, size=(160,))
w = df['Stim timestamp'] + np.random.uniform(low=-0.1, high=0.75, size=(160,))
# random.normal(loc=0.6, scale=0.80, size=None)
qq = np.abs(q - w)
```

```python
df['key timestamp'] = q
df['Stim timestamp'] = w
df['delta T'] = qq
df[:5]
```

```python
df.to_csv('/trial_2.csv')
df = pd.read_csv('/trial_2.csv')
df[:5]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))

ax.hist(df[df['is congruent'] == 1]['delta T'], alpha=0.5, color='r', label='Congruent', bins=16, density=True)
ax.hist(df[df['is congruent'] == 0]['delta T'], alpha=0.5, color='g', label='Incongruent', bins=16, density=True)
ax.legend()
ax.set_title("Congurent & Incongruent", fontsize=19)
plt.grid(alpha=0.2)
fig.savefig('/Congurent_Incongruent_hist.png')
plt.show
```
